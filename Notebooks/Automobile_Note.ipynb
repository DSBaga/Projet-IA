```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
## Importation des données
df = pd.read_csv('../Data/Automobile_data.csv')
df.head()
```

```python
# Description des données
df.describe()
```

```python
# Affichage des colonnes et leurs types
df.info()
```

```python

df['price']= pd.to_numeric(df['price'], errors='coerce')
```

```python
df.info()
```

```python
#sns.pairplot(df)
```

```python
## choisir uniquement les variables numériques
#df = df.select_dtypes(include=[np.number])
df_num = df.select_dtypes(exclude='object')
df_num.head()
```

```python
df_num.columns  
```

```python
df_num = df_num.dropna()
```

```python
df_num.shape
```

```python
# features
X = df_num.drop('price', axis=1)
# target
y = df_num['price']

# train test split
seed = 123
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
print (X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
## Modèle de régression linéaire multiple
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
def evaluate(model) :
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)
    rmse_train = mean_squared_error(y_train, train_preds, squared=False)
    rmse_test = mean_squared_error(y_test, test_preds, squared=False)
    return rmse_train, rmse_test
```

```python
evaluate(lr)
```

```python
## Model de random forest
rf = RandomForestRegressor(random_state=seed)
rf.fit(X_train, y_train)
```

```python
evaluate(rf)
```

```python
# Importation du modele avec joblib
import joblib
joblib.dump(rf, '../Models/Automibile_final_model.joblib')
```
